```python
import yfinance as yf

def fetch_market_cap_on_date(stock_symbol, date):
    # Fetch historical stock prices around the specified date
    stock = yf.Ticker(stock_symbol)
    # Fetching 2 days to ensure we get data in case one of the dates falls on a weekend/holiday
    hist = stock.history(start=pd.to_datetime(date) - pd.Timedelta(days=1), end=pd.to_datetime(date) + pd.Timedelta(days=1))
    
    # Attempt to fetch the most recent number of shares outstanding
    shares_outstanding = stock.info.get('sharesOutstanding')
    
    if shares_outstanding is None:
        print("Shares outstanding information is not available.")
        return None

    # Get the close price for the specified date (if available)
    if date in hist.index:
        close_price = hist.loc[date]['Close']
    else:
        print("Close price not available for the specified date. Please check the date and try again.")
        return None
    
    # Calculate the approximate market cap
    market_cap = close_price * shares_outstanding
    
    return market_cap
```

```python
market_cap_dfs = {}
```

```python
def fetch_historical_market_cap(stock_symbol, start_date, end_date):
    stock = yf.Ticker(stock_symbol)
    hist = stock.history(start=start_date, end=end_date)
    
    shares_outstanding = stock.info.get('sharesOutstanding')
    
    if shares_outstanding is None:
        print("Shares outstanding information is not available.")
        return None

    hist['Shares Outstanding'] = shares_outstanding
    
    return hist[['Shares Outstanding']]

for ticker, range_list in ticker_range_map.items():
    current_df = fetch_historical_market_cap(ticker, range_list[0], range_list[1])
    market_cap_dfs[ticker] = current_df
```

```python
for ticker, df in market_cap_dfs.items():
    # Check if df is not None before attempting assignment
    if df is not None:
        df['Ticker_Symbol'] = ticker
    else:
        print(f"Warning: No DataFrame found for ticker {ticker}.")
```
