## Hypothesis Test: There exists NO correlation between the income of an adult and an unhealthy lifestyle.
<p> Unhealthy lifestyle determined by lack of exercise, obesity/overweight, and food consumption patterns (eg. eating fruits/veg less than once daily)</p>

```python
question_ids = ['Q047', 'Q036', 'Q037', 'Q018', 'Q019']
obesity_related_df = obesity_df[obesity_df['QuestionID'].isin(question_ids)]
obesity_related_df['Income'].isna().sum()
```

```python
obesity_related_df.dropna(subset=['Income'], inplace=True)
obesity_related_df
```

```python
obesity_related_df = obesity_related_df[obesity_related_df['Income'] != 'Data not reported']
obesity_related_df
```

```python
from pprint import pprint
#categorical column mapping: ordinal income column
income_mapping = {'Less than $15,000': 0,
'$15,000 - $24,999': 1,
 '$25,000 - $34,999': 2,
 '$35,000 - $49,999': 3,
 '$50,000 - $74,999': 4,
 '$75,000 or greater': 5}
pprint(income_mapping)
```

```python
from scipy.stats import spearmanr
obesity_related_df['Income_Numeric'] = obesity_related_df['Income'].map(income_mapping)
obesity_related_df
```

```python
obesity_related_df.dropna(subset=['Data_Value'], inplace=True)
```

```python
obesity_related_df.isna().sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  
plt.scatter(obesity_related_df['Income'], obesity_related_df['Data_Value'],)  
plt.title('Relationship between Income and Unhealthy Lifestyle Measure')  
plt.xlabel('Numeric Income')  
plt.ylabel('Unhealthy Lifestyle Measure (Data Value)')  
plt.grid(True)  
plt.show()


# Calculate Spearman correlation
correlation, p_value = spearmanr(obesity_related_df['Income_Numeric'], obesity_related_df['Data_Value'])

print("Spearman Correlation Coefficient:", correlation)
print("P-value:", p_value) #there exists a weak and NEGATIVE correlation
#null hypothesis can be rejected
```

```python

def merge_with_unique_cols(df_left, df_right, on_keys, suffixes=('_left', '_right')):
    df_right_renamed = df_right.rename(columns={
        col: f"{col}{suffixes[1]}" if col not in on_keys else col 
        for col in df_right.columns
    })
    return pd.merge(df_left, df_right_renamed, on=on_keys, how='outer')

industry_related_df = merge_with_unique_cols(meat_prod_df, cold_storage_df, on_keys=['Date', 'Year', 'Month'])
industry_related_df = merge_with_unique_cols(industry_related_df, sl_count_df, on_keys=['Date', 'Year', 'Month'])
industry_related_df = merge_with_unique_cols(industry_related_df, sl_weights_df, on_keys=['Date', 'Year', 'Month'])


industry_related_df
```
