```python
# Get KO data from Yahoo Finance
KO = yf.Ticker("KO")
KO_data = KO.history(start="2000-01-01")

KO_data
```

```python
KO_data = KO_data.reset_index()
KO_data["Date"] = pd.to_datetime(KO_data["Date"])

KO_data = KO_data[["Date", "Close"]]

KO_data
```

```python
# Assuming 'KO_data' is your DataFrame and it's indexed by date
KO_data.index = pd.to_datetime(KO_data["Date"])

# Resample to get the last trading day of each month
monthly_close = KO_data.resample('M').last()


monthly_close
```

```python
# Drop the time from the index 
monthly_close.index = monthly_close.index.date

monthly_close

# Drop the Date column
monthly_close = monthly_close.drop(columns=["Date"])

monthly_close.reset_index(inplace=True)
```

```python
monthly_close.rename(columns={"index": "Date"}, inplace=True)

monthly_close
```

```python
all_data = pd.read_csv('Data/all_stock_and_etfs.csv')

all_data
```

```python
# Only get the data where the Ticker_Symbol is SPY, reset index and have it in ascending order by Date-Time
SPY_data = all_data[all_data["Ticker_Symbol"] == "SPY"]
SPY_data = SPY_data.reset_index(drop=True)
SPY_data = SPY_data.sort_values(by="Date-Time", ascending=True)

# Only get data from 2000-01-01
SPY_data = SPY_data[SPY_data["Date-Time"] >= "2000-01-01"]

SPY_data.index = pd.to_datetime(SPY_data["Date-Time"])

# Resample to get the last trading day of each month
monthly_close_SPY = SPY_data.resample('M').last()

# Drop the time from the index
monthly_close_SPY.index = monthly_close_SPY.index.date

# Drop the Date-Time column
monthly_close_SPY = monthly_close_SPY.drop(columns=["Date-Time"])

monthly_close_SPY.reset_index(inplace=True)

monthly_close_SPY.rename(columns={"index": "Date"}, inplace=True)

monthly_close_SPY = monthly_close_SPY[["Date", "Close"]]

monthly_close_SPY
```

```python
# Rename the Close columns in both to Ticker_Close
monthly_close.rename(columns={"Close": "KO_Close"}, inplace=True)
monthly_close_SPY.rename(columns={"Close": "SPY_Close"}, inplace=True)

# Get the log returns for both
monthly_close["KO_Log_Return"] = np.log(
    monthly_close["KO_Close"]) - np.log(monthly_close["KO_Close"].shift(1))

monthly_close_SPY["SPY_Log_Return"] = np.log(
    monthly_close_SPY["SPY_Close"]) - np.log(monthly_close_SPY["SPY_Close"].shift(1))

# Merge the two DataFrames on Date
merged_df = pd.merge(monthly_close, monthly_close_SPY, on="Date")

merged_df
```

```python
# Graph both the KO_Log_Return and SPY_Log_Return

plt.plot(merged_df["Date"], merged_df["KO_Log_Return"], label="KO")
plt.plot(merged_df["Date"], merged_df["SPY_Log_Return"], label="SPY")
```

```python
# Generate a column for excess returns
merged_df["Excess_Return"] = merged_df["KO_Log_Return"] - merged_df["SPY_Log_Return"]

# Graph the Excess_Return
plt.plot(merged_df["Date"], merged_df["Excess_Return"], label="Excess Return")
```

```python
all_commodities = pd.read_csv('Data/all_commodities.csv')

# Get all the data for Commodity = Sugar, reset the index and have it in ascending order by Date
sugar_price_data = all_commodities[all_commodities["Commodity"] == "Sugar"]

sugar_price_data["Date-Time"] = pd.to_datetime(sugar_price_data["Date-Time"])

# Rename the Date-Time column to Date
sugar_price_data.rename(columns={"Date-Time": "Date"}, inplace=True)

sugar_price_data = sugar_price_data.sort_values(by="Date", ascending=True)

sugar_price_data = sugar_price_data.reset_index(drop=True)

# Rename the Value column to Sugar_Price, Drop the Commodity column
sugar_price_data.rename(columns={"Value": "Sugar_Price"}, inplace=True)

# Get the log returns for the Sugar_Price
sugar_price_data["Sugar_Log_Return"] = np.log(
    sugar_price_data["Sugar_Price"]) - np.log(sugar_price_data["Sugar_Price"].shift(1))

# Drop the Commodity Column
sugar_price_data = sugar_price_data.drop(columns=["Commodity", "Unit"])

sugar_price_data
```
