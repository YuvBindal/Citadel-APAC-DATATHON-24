```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Open Meat_Stats_Meat_Production.csv
meat_production_df = pd.read_csv('Data/Meat_Stats_Meat_Production.csv')

display(meat_production_df.head())
meat_production_df.info()
```

```python
meat_production_df["Date"] = pd.to_datetime(meat_production_df["Date"], format='%b-%Y')

display(meat_production_df.head())
```

```python
# Find me rows in the 
```

```python
# Drop rows with NaN values
meat_production_df = meat_production_df.dropna()

meat_production_df.info()
```

```python
# Unique values in Animal column
meat_production_df['Animal'].unique()

# Unique values in Type of Meat column
meat_production_df['Type of Meat'].unique()
```

```python
# Create a df for 2000 data onwards
meat_production_df_2000 = meat_production_df[meat_production_df['Year'] >= 2000]

display(meat_production_df_2000)
```

```python
# Create a df for 2000 data onwards for Red Meat only from the Type of Meat column
meat_production_df_2000_red_meat = meat_production_df_2000[meat_production_df_2000['Type of Meat'] == 'Red Meat']

display(meat_production_df_2000_red_meat)
```

```python
# Remove commas and convert to float
meat_production_df_2000_red_meat['Production'] = meat_production_df_2000_red_meat['Production'].str.replace(
    ',', '').astype(float)

# Now perform the groupby operation
meat_production_df_2000_red_meat_sum = meat_production_df_2000_red_meat.groupby(
    ['Date', 'Animal', 'Unit', 'Type of Meat', 'Year', 'Month']).sum().reset_index()

meat_production_df_2000_red_meat_sum.drop(columns=['Commercial or Federally Inspected'], inplace=True)
```

```python
display(meat_production_df_2000_red_meat_sum)
```

```python
# For each Animal, plot the Production vs Date. Since the Production values are different for each Animal, we will plot on different axes
fig, ax = plt.subplots(meat_production_df_2000_red_meat_sum['Animal'].nunique(), 1, figsize=(10, 10))

for i, animal in enumerate(meat_production_df_2000_red_meat_sum['Animal'].unique()):
    animal_df = meat_production_df_2000_red_meat_sum[meat_production_df_2000_red_meat_sum['Animal'] == animal]
    ax[i].plot(animal_df['Date'], animal_df['Production'])
    ax[i].set_title(animal)
    ax[i].set_xlabel('Date')
    ax[i].set_ylabel('Production')
    ax[i].grid()

plt.tight_layout()
plt.show()
```

```python
meat_production_df_2000_red_meat_grouped_sum = meat_production_df_2000_red_meat_sum.groupby(
    ['Date', 'Unit', 'Type of Meat', 'Year', 'Month']).sum().reset_index()

meat_production_df_2000_red_meat_grouped_sum.drop(columns=['Animal'], inplace=True)

display(meat_production_df_2000_red_meat_grouped_sum)
```

```python
# Graph for Red Meat Production vs Date
plt.figure(figsize=(10, 5))
plt.plot(meat_production_df_2000_red_meat_grouped_sum['Date'],
         meat_production_df_2000_red_meat_grouped_sum['Production'])
plt.title('Red Meat Production vs Date')
plt.xlabel('Date')
plt.ylabel('Production')
plt.grid()
```

```python
# Use the above data frame to get yearly data
meat_production_df_2000_red_meat_grouped_sum_yearly = meat_production_df_2000_red_meat_grouped_sum.drop(columns=['Month', 'Date'])

meat_production_df_2000_red_meat_grouped_sum_yearly = meat_production_df_2000_red_meat_grouped_sum_yearly.groupby(
    ['Year', 'Unit', 'Type of Meat']).sum().reset_index()

display(meat_production_df_2000_red_meat_grouped_sum_yearly)
```

```python
# Graph for Red Meat Production vs Year
plt.figure(figsize=(10, 5))
plt.plot(meat_production_df_2000_red_meat_grouped_sum_yearly['Year'],
         meat_production_df_2000_red_meat_grouped_sum_yearly['Production'])
plt.title('Red Meat Production vs Year')
plt.xlabel('Year')
plt.ylabel('Production')
plt.grid()

plt.show()
```
